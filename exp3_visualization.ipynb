```python
import matplotlib.pyplot as plt
import numpy as np
import pickle

plt.rcParams['font.family'] = 'serif'
plt.rcParams.update({"text.usetex": True,
                        "text.latex.preamble": r"\usepackage{amsmath}"})
plt.rcParams.update({'pdf.fonttype': 42})
```

```python
# load data
folder = "exp3_results"
exp_no_circ = "no_circ_2024-08-23-15-56-21"
file = 'data.pickle'
with open(folder + '/' + exp_no_circ + '/' + file, 'rb') as f:
    data_no_circ = pickle.load(f)
```

```python
print(data_no_circ.keys())
```

```python
t = np.cumsum(data_no_circ['time_per_loop'])
dq = np.array(data_no_circ["dq"])
dq_norm = np.linalg.norm(dq, axis=1)
start_idx = np.where((dq_norm > 0.05) & (t > 11))[0][0]
end_idx = np.where((dq_norm > 0.05) & (t > 11))[0][-1]
```

```python
all_h = np.array(data_no_circ['all_h'])
print(all_h.shape)
```

```python
fig, ax = plt.subplots(figsize=(5,4), frameon=True)
label_fs = 20
tick_fs = 20
legend_fs = 20
linewidth = 3

plt.plot(t, np.min(all_h, axis=1), linewidth=linewidth)
plt.hlines(0, t[0], t[-1], colors='k', linestyles='dashed', linewidth=linewidth)
plt.vlines(t[start_idx], np.min(np.min(all_h, axis=1)), np.max(np.min(all_h, axis=1)), colors='g', linestyles='dashed')
plt.vlines(t[end_idx], np.min(np.min(all_h, axis=1)), np.max(np.min(all_h, axis=1)), colors='r', linestyles='dashed')
```

```python
h_picked = all_h[start_idx:end_idx,:2]
t_picked = t[start_idx:end_idx] - t[start_idx]

fig, ax = plt.subplots(figsize=(5,4), frameon=True)
label_fs = 20
tick_fs = 20
legend_fs = 20
linewidth = 3

plt.plot(t_picked, np.min(h_picked, axis=1), linewidth=linewidth)
plt.hlines(0, t_picked[0], t_picked[-1], colors='k', linestyles='dashed', linewidth=linewidth)

plt.xlabel('$t$', fontsize=label_fs)
plt.ylabel('$\min(h_1, ..., h_K)$', fontsize=label_fs)
plt.xticks(fontsize = tick_fs)
plt.yticks(fontsize = tick_fs)
plt.tight_layout()
plt.savefig(folder + '/' + exp_no_circ + '/cbf_no_circ.pdf')
```
