```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
# load data
folder = "exp1_results"
exp = "no_circ_2024-07-22-12-06-21"
file = 'data.pickle'
with open(folder + '/' + exp + '/' + file, 'rb') as f:
    data_no_circ = pickle.load(f)
```

```python
print(data_no_circ.keys())
```

```python
all_h = np.array(data_no_circ['all_h'])
plt.plot(np.min(all_h, axis=1))
plt.hlines(0, 0, len(all_h), colors='r', linestyles='dashed')
```

```python
# load data
folder = "exp1_results"
exp = "with_circ_2024-07-22-12-17-43"
file = 'data.pickle'
with open(folder + '/' + exp + '/' + file, 'rb') as f:
    data_with_circ = pickle.load(f)
```

```python
print(data_with_circ.keys())
```

```python
smooth_min = np.array(data_with_circ['smooth_min'])
t = np.cumsum(data_with_circ['time_per_loop'])
plt.plot(t, smooth_min)
plt.hlines(0, t[0], t[-1], colors='r', linestyles='dashed')
```
