```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
# load data
folder = 'exp1_results'
file = 'data_no_circ.pickle'
with open(folder + '/' + file, 'rb') as f:
    data_no_circ = pickle.load(f)
```

```python
print(data_no_circ.keys())
```

```python
all_h = np.array(data_no_circ['all_h_list'])
plt.plot(np.min(all_h, axis=1))
plt.hlines(0, 0, len(all_h), colors='r', linestyles='dashed')
```
